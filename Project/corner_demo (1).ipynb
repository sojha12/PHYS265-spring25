# corner.py Demonstration: Visualizing Multivariate Posterior Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
!pip install corner 
import corner as corner
```

```python
ndim = 6
means = np.array([0.0, 0.5, -0.5, 1.0, -1.0, 0.25])
cov = 0.3 * np.ones((ndim, ndim)) + 0.7 * np.eye(ndim)

np.random.seed(123)
samples = np.random.multivariate_normal(means, cov, size=15000)
```

```python
labels = ["x1", "x2", "x3", "x4", "x5", "x6"]
figure = corner.corner(samples, labels=labels, truths=means,
                       show_titles=True, title_fmt=".2f", title_kwargs={"fontsize": 12})
plt.suptitle("Corner Plot of 6D Multivariate Samples")
plt.show()
```

```python
corner.corner(samples[:, :3], labels=labels[:3], truths=means[:3],
              show_titles=True, title_fmt=".2f", title_kwargs={"fontsize": 12})
plt.suptitle("Corner Plot of First 3 Dimensions")
plt.show()
```

### Figure Explanations

- **6D Plot**: Full multivariate normal samples showing pairwise correlations and 1D histograms for all six parameters.
- **3D Plot**: Focused on first three parameters for detailed view.

Generated using [`corner.py`](https://github.com/dfm/corner.py)  a popular tool in astrophysics and Bayesian analysis.
